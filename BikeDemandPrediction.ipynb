```python
#Assigning string value to different months
def object_map_mnth(x):
    return x.map({1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'})
```

```python
dataset[['mnth']]=dataset[['mnth']].apply(object_map_mnth)
```

```python
def str_map_weekday(x):
    return x.map({1:'Mon',2:'Tue',3:'Wed',4:'Thu',5:'Fri',6:'Sat',0:'Sun'})
```

```python
dataset[['weekday']]=dataset[['weekday']].apply(str_map_weekday)
```

# #Data Preparion

```python
dataset_categorical=dataset.select_dtypes(include=['object'])
```

```python
dataset_categorical.head()
```

```python
dataset_dumies=pd.get_dummies(dataset_categorical,drop_first=True)
dataset_dumies.head()
```

```python
#Drop Categorical variablel column
dataset=dataset.drop(list(dataset_categorical.columns), axis=1)
dataset
```

```python
#Concatenate dummy variable with the dataset
dataset=pd.concat([dataset,dataset_dumies], axis=1)
```

```python
dataset.head()
```

```python
# The drop method returns a new DataFrame with the specified columns removed
dataset = dataset.drop(['instant', 'dteday'], axis=1)

# Now 'dataset' holds the modified DataFrame
dataset.head()
```
